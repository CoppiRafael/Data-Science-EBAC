#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
yf_mavg = yf['Adj Close'].rolling(window=90).mean()
yf_mavg
````

````python
yf_mstd = yf['Adj Close'].rolling(window=90).std()
yf_mstd
````

````python
yf_mavg_stack = yf_mavg.stack().reset_index()
yf_mavg_stack.columns = ['Data','Ativo','Valor']

fig = px.line(data_frame=yf_mavg_stack,x='Data',y='Valor',color='Ativo',template='plotly_dark')
fig.show()
````

````python
yf_STD_stack = yf_mstd.stack().reset_index()
yf_STD_stack.columns = ['Data','Ativo','Valor']

fig = px.line(data_frame=yf_STD_stack,x='Data',y='Valor',color='Ativo',template='plotly_dark')
fig.show()
````

---

### Analisando


**Podemos ver que claramente o gráfico da média é bem mais volátil, principalmente quando observamos as 'Highs and Lows', é bem nítido que o desvio padrão contém mais mais oscilação.Mas em resumo, elas se convergem no quesito geral, na qual ambas passam por um momento de tendência de baixa, porém a média começa cair a parti de dezembro de 2021, já o desvio padrão começa a demontrar fraqueza ja em Fevereiro de 2021. Podemos observa uma divergência das métricas(linha verde), na qual uma renova o topo e outra não. Também o mmomento de tendência de baixa para o desvio é bem mais agudo e volátil do que para a média. Posteriormente ambas começam a subir porém a média com uma alta bem mais satisfatória.**

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df = yf.stack(future_stack=True).reset_index()
df = df.loc[df['Ticker'] == 'BTC-USD',:]
````

````python
df_tail_60 = df.tail(60)
df_tail_60.shape
````

````python
df_tail_60
````

````python
fig = go.Figure(data=[go.Candlestick(x=df_tail_60['Date'],
                                open = df_tail_60['Open'],
                                high =df_tail_60['High'],
                                low = df_tail_60['Low'],
                                close = df_tail_60['Close'],
                                increasing_line_color= 'green',
                                decreasing_line_color= 'gray',
                                    )])
fig.show()
````
