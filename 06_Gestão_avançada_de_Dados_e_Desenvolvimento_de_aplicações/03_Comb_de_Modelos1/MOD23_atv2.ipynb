# Tarefa 2: Random Forest

### 1. Passo a Passo para o Algoritmo Random Forest

1. Criação de Amostras Bootstrap:

    - A partir do seu dataset original, crie várias amostras bootstrap. Isso significa que você vai criar conjuntos de dados de treino escolhendo aleatoriamente, com reposição, várias linhas do dataset original. Se você tiver 100 linhas, cada amostra bootstrap também terá 100 linhas, mas algumas linhas serão repetidas.
---
2. Seleção Aleatória de Colunas:

    - Quando for dividir um nó na árvore, o Random Forest escolhe aleatoriamente um subconjunto de variáveis em vez de considerar todas. Isso significa que, se você tiver 10 colunas, talvez a árvore só considere 3 delas para tomar uma decisão em cada nó.
---
3. Treinamento de Árvores de Decisão:

    - Para cada amostra bootstrap, uma árvore de decisão é treinada. Como cada árvore será treinada com uma amostra diferente e usando diferentes colunas a cada divisão, as árvores serão diversificadas.

---
4. Previsão:

    - Quando você quer prever um novo dado, cada árvore na floresta faz sua previsão.
    - Para classificação: A classe com mais votos entre todas as árvores é a predição final.
    - Para regressão: A média das predições de todas as árvores é a predição final.

---
5. Agregação dos Resultados:

    - Combine os resultados de todas as árvores para fazer uma predição final. Isso reduz o overfitting e melhora a precisão do modelo.

### 2. Explicação Simples sobre Random Forest
O Random Forest é uma coleção de várias árvores de decisão que são treinadas com diferentes partes dos dados e diferentes combinações de variáveis. Ele melhora a precisão e a robustez das predições, porque usa uma média ou votação entre muitas árvores para fazer previsões. A ideia principal é que, ao combinar muitos modelos diferentes (as árvores), o Random Forest evita erros que um modelo único poderia cometer, tornando a predição final mais confiável.

### 3. Diferença entre Bagging e Random Forest
- No bagging, várias amostras bootstrap são usadas para treinar árvores de decisão. No entanto, todas as árvores usam todas as colunas (variáveis) disponíveis para fazer suas divisões.
- No Random Forest, além de usar amostras bootstrap como no bagging, ele também introduz a ideia de seleção aleatória de colunas a cada divisão de nó. Isso aumenta a diversidade entre as árvores, tornando o modelo ainda mais robusto e reduzindo a correlação entre as árvores.


Resumindo, o Random Forest é basicamente bagging com uma "pitada" de aleatoriedade extra, fazendo com que as árvores sejam mais diferentes entre si.

### 4. Implementação em Python

```python
# Importando bibliotecas necessárias
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

# Gerando um dataset fictício para exemplo
X, y = make_classification(n_samples=100, n_features=10, n_informative=8, n_classes=2, random_state=42)

# Dividindo o dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando o modelo de Random Forest
rf = RandomForestClassifier(n_estimators=100, max_features='sqrt', random_state=42)

# Treinando o modelo
rf.fit(X_train, y_train)

# Fazendo predições
y_pred = rf.predict(X_test)

# Avaliando o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo Random Forest: {accuracy:.2f}")
```
