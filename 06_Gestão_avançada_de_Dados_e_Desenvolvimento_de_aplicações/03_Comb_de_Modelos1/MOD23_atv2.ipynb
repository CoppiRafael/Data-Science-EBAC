# Tarefa 2: Random Forest

### 1. Passo a Passo para o Algoritmo Random Forest

1. Criao de Amostras Bootstrap:

    - A partir do seu dataset original, crie vrias amostras bootstrap. Isso significa que voc vai criar conjuntos de dados de treino escolhendo aleatoriamente, com reposio, vrias linhas do dataset original. Se voc tiver 100 linhas, cada amostra bootstrap tambm ter 100 linhas, mas algumas linhas sero repetidas.
---
2. Seleo Aleatria de Colunas:

    - Quando for dividir um n na rvore, o Random Forest escolhe aleatoriamente um subconjunto de variveis em vez de considerar todas. Isso significa que, se voc tiver 10 colunas, talvez a rvore s considere 3 delas para tomar uma deciso em cada n.
---
3. Treinamento de rvores de Deciso:

    - Para cada amostra bootstrap, uma rvore de deciso  treinada. Como cada rvore ser treinada com uma amostra diferente e usando diferentes colunas a cada diviso, as rvores sero diversificadas.

---
4. Previso:

    - Quando voc quer prever um novo dado, cada rvore na floresta faz sua previso.
    - Para classificao: A classe com mais votos entre todas as rvores  a predio final.
    - Para regresso: A mdia das predies de todas as rvores  a predio final.

---
5. Agregao dos Resultados:

    - Combine os resultados de todas as rvores para fazer uma predio final. Isso reduz o overfitting e melhora a preciso do modelo.

### 2. Explicao Simples sobre Random Forest
O Random Forest  uma coleo de vrias rvores de deciso que so treinadas com diferentes partes dos dados e diferentes combinaes de variveis. Ele melhora a preciso e a robustez das predies, porque usa uma mdia ou votao entre muitas rvores para fazer previses. A ideia principal  que, ao combinar muitos modelos diferentes (as rvores), o Random Forest evita erros que um modelo nico poderia cometer, tornando a predio final mais confivel.

### 3. Diferena entre Bagging e Random Forest
- No bagging, vrias amostras bootstrap so usadas para treinar rvores de deciso. No entanto, todas as rvores usam todas as colunas (variveis) disponveis para fazer suas divises.
- No Random Forest, alm de usar amostras bootstrap como no bagging, ele tambm introduz a ideia de seleo aleatria de colunas a cada diviso de n. Isso aumenta a diversidade entre as rvores, tornando o modelo ainda mais robusto e reduzindo a correlao entre as rvores.


Resumindo, o Random Forest  basicamente bagging com uma "pitada" de aleatoriedade extra, fazendo com que as rvores sejam mais diferentes entre si.

### 4. Implementao em Python

```python
# Importando bibliotecas necessrias
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

# Gerando um dataset fictcio para exemplo
X, y = make_classification(n_samples=100, n_features=10, n_informative=8, n_classes=2, random_state=42)

# Dividindo o dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando o modelo de Random Forest
rf = RandomForestClassifier(n_estimators=100, max_features='sqrt', random_state=42)

# Treinando o modelo
rf.fit(X_train, y_train)

# Fazendo predies
y_pred = rf.predict(X_test)

# Avaliando o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurcia do modelo Random Forest: {accuracy:.2f}")
```
