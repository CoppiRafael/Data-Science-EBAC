1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

Definindo os parmetros da distribuio normal:

- Mdia () = 15 kg
- Desvio padro () = 0,1 kg

Calculando a probabilidade:
- Precisamos calcular P(X>15.2)
- Isso pode ser feito utilizando a CDF da distribuio normal. A probabilidade de X ser maior que 15.2 kg  igual a 1 - P(X  15.2)



```python
mu = 15  # mdia
sigma = 0.1  # desvio padro

# Peso alvo
peso_alvo = 15.2

# Calculando a probabilidade
probabilidade_maior_que_15_2 = 1 - stats.norm.cdf(peso_alvo, loc=mu, scale=sigma)

print(f'A probabilidade de um pacote ter mais que 15,2 kg : {(probabilidade_maior_que_15_2)*100:.4f} %')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
mu = 15  
sigma = 0.1  

# Calculando o percentil de 95%
percentil_95 = stats.norm.ppf(0.95, loc=mu, scale=sigma)

print(f'O percentil de ordem 95% para o peso dos pacotes : {percentil_95:.4f} kg')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades para a distribuio normal padro
probabilidade_normal = stats.norm.cdf(Y)

# Calculando as probabilidades para a distribuio t-Student com 5 graus de liberdade
probabilidade_t = stats.t.cdf(Y, df=5)

# Exibindo as probabilidades
print("Probabilidades para Normal Padro:")
for y, p in zip(Y, probabilidade_normal):
    print(f"P(X < {y:.2f}) = {p:.4f}")

print("\nProbabilidades para t-Student (df=5):")
for y, p in zip(Y, probabilidade_t):
    print(f"P(T < {y:.2f}) = {p:.4f}")
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Criando o vetor Y
Y = np.linspace(-6, 6, 200)

# Calculando as probabilidades para a distribuio normal padro
probabilidade_normal = stats.norm.cdf(Y)

# Calculando as probabilidades para a distribuio t-Student com 5 graus de liberdade
probabilidade_t = stats.t.cdf(Y, df=5)

# Visualizando os resultados
plt.figure(figsize=(12, 6))


plt.plot(Y, probabilidade_normal, label='Normal Padro', color='blue', linewidth=2)
plt.plot(Y, probabilidade_t, label='t-Student (df=5)', color='orange', linewidth=2)

plt.title('Distribuio das Probabilidades Acumuladas')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')
plt.axhline(0.95, color='red', linestyle='--', label='95% de Probabilidade')
plt.axvline(0, color='gray', linestyle='--')  
plt.legend()
plt.grid()
plt.xlim(-6, 6)  
plt.ylim(0, 1)   
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Criando o vetor Y com 20 valores entre 0 e 1
Y = np.linspace(0, 1, 20)

percentis_normal = stats.norm.ppf(Y)

percentis_t = stats.t.ppf(Y, df=5)

print("Percentis para a Normal Padro:")
for y, p in zip(Y, percentis_normal):
    print(f"P(Normal Padro) para Y={y:.2f} : {p:.4f}")

print("\nPercentis para t-Student (df=5):")
for y, p in zip(Y, percentis_t):
    print(f"P(t-Student) para Y={y:.2f} : {p:.4f}")
```

6. Faa um grfico com os dados do item anterior.

```python

Y = np.linspace(0, 1, 20)

# Calculando os percentis para a distribuio normal padro
percentis_normal = stats.norm.ppf(Y)

# Calculando os percentis para a distribuio t-Student com 5 graus de liberdade
percentis_t = stats.t.ppf(Y, df=5)

# Criando o grfico
plt.figure(figsize=(12, 6))
plt.plot(Y, percentis_normal, marker='o', label='Percentis - Normal Padro', color='blue')
plt.plot(Y, percentis_t, marker='o', label='Percentis - t-Student (df=5)', color='orange')

# Adicionando detalhes ao grfico
plt.title('Percentis para Distribuies Normal Padro e t-Student')
plt.xlabel('Valores de Y (Probabilidade Acumulada)')
plt.ylabel('Percentis')
plt.axhline(0, color='gray', linestyle='--')  
plt.axvline(0, color='gray', linestyle='--')  
plt.legend()
plt.grid()
plt.xlim(0, 1) 
plt.ylim(-4, 4)  
plt.show()
```
