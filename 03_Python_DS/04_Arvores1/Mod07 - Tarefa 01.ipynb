*Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.*

````python
df_variav = pd.DataFrame(df.dtypes).reset_index()
df_variav.rename(columns={'index':'variavel',0:'type'},inplace=True)
````

Adicione uma coluna nesse dataframe chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 

````python
new_column = list()

for coluna in df.columns.to_list():
    new_column.append(df[coluna].nunique())

df_variav['qtd_categorias'] = new_column
df_variav
````

Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
#inicialmente tratando essas variáves contínuas. Pq não podem ser do tipo 'Float'
df.idade = df.idade.astype(int)
df.qt_pessoas_residencia = df.qt_pessoas_residencia.astype(int)
df.mau = df.mau.astype(int)
df.info();df.head()
````

````python
df['posse_de_veiculo'] = df['posse_de_veiculo'].map({'Y':1,'N':0})
df['posse_de_imovel'] = df['posse_de_imovel'].map({'Y':1,'N':0})
df['sexo'] = df['sexo'].map({'M':0,'F':1})
````

````python
df_base = pd.get_dummies(df,['tipo_renda','estado_civil','educacao','tipo_residencia'])
display(df.head())
display(df_base.head())
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tab = pd.crosstab(df_base['possui_email'],df_base['posse_de_veiculo'])
tab.div(tab.sum(axis=1),axis=0)
````

````python
pd.crosstab(df_base['possui_email'],df_base['mau']).div(tab.sum(axis=1),axis=0)
````

````python
pd.crosstab(df_base['posse_de_veiculo'],df_base['mau']).div(tab.sum(axis=1),axis=0)
````
