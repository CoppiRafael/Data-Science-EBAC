#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sinasc_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df_copia = df.loc[:,[ 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO','ESTCIVMAE','CONSULTAS']]
```

```python
fig,axes = plt.subplots(nrows=2,ncols=3,figsize=(10,8))
sns.set_theme(style="whitegrid", palette="pastel")

sns.countplot(ax=axes[0][0],x = 'APGAR1', data=df_copia)
sns.countplot(ax=axes[0][1],x = 'APGAR5', data=df_copia)
sns.countplot(ax=axes[0][2],x = 'SEXO', data=df_copia)
sns.countplot(ax=axes[1][0],x = 'GESTACAO',data=df_copia)
sns.countplot(ax=axes[1][1],x = 'ESTCIVMAE',data=df_copia)

# Rotacionando os ticks do eixo x
axes[1][0].set_xticklabels(axes[1][0].get_xticklabels(), rotation=75)
axes[0][0].set_xticklabels(axes[0][0].get_xticklabels(), rotation=45)
axes[0][1].set_xticklabels(axes[0][1].get_xticklabels(), rotation=45)
axes[1][0].set_xticklabels(axes[1][0].get_xticklabels(), rotation=45)
axes[1][1].set_xticklabels(axes[1][1].get_xticklabels(), rotation=45)
axes[0][2].set_xticklabels(axes[0][2].get_xticklabels(), rotation=45)
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(10,5))
sns.set_theme(style="whitegrid", palette="pastel")

sns.barplot( x = 'CONSULTAS', y='APGAR5', data=df_copia)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(10,8))
sns.set_theme(style="whitegrid", palette="pastel")

sns.countplot(ax=axes[0][0],x = 'SEXO', data=df_copia)
sns.countplot(ax=axes[0][1],x = 'GESTACAO',data=df_copia)
sns.countplot(ax=axes[1][0],x = 'ESTCIVMAE', data=df_copia)
sns.countplot(ax=axes[1][1],x = 'CONSULTAS',data=df_copia)

axes[0][1].set_xticklabels(axes[0][1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```
