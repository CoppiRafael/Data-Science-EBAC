#### Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), e que devolve os dados da consulta no formato DataFrame.

````python
#1) Seu código aqui
def estado(tipo:str=None,uf:str=None):
    
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo}'
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data['items'])
        return df
    else:
        print(f'Erro ao fazer request. Erro:{response.status_code}')
        return

#Testando
uf = 'RJ'
tipo_interessado = 'Estado'
df_teste = estado(tipo_interessado,uf)
df_teste.head()
````

````python
df = pd.DataFrame(data=requests.get('http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl').json()['items'])
````

#### Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?

**Qual é o município da Bahia com mais solicitações deferidas?**


````python
# 3) Seu código aqui
for k,v in df[(df['uf'] == 'BA') & (df['tipo_interessado']=='Município')]['interessado'].value_counts().to_dict().items():
    print(f"Município: {k} , Quantidade de solicitações: \033[91m{v}\033[0m")
````
