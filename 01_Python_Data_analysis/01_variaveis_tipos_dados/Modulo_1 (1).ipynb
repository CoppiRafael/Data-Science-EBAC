2.   Item da lista




```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

#- 1
print(cancao.title())
print('-='*25)
#- 32
print(cancao.lower())
print('-='*25)
#- 3
print(cancao.upper())
print('-='*25)
#- 4
cancao = cancao.title().split(',')
for x,v in enumerate(cancao):
  print(f'{x} : {v}')
```

#Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

3.   Item da lista


```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

selic = noticia[noticia.find('2,75%') : (noticia.find('2,75%') + 5) ]
print(f'{selic}')
print('-='*3)
ano = noticia[noticia.find('6'):(noticia.find('6') + 1)]
print(ano)
```
