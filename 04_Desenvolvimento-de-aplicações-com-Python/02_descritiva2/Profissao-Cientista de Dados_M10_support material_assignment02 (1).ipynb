````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df.tail()
````

````python
df[['renda','posse_de_imovel','posse_de_veiculo']].isna().sum()
````

````python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(15,8))
sns.countplot(data=df, x='posse_de_imovel',palette='Blues',ax=ax[0][0])
sns.countplot(data=df, x='posse_de_veiculo',palette='Blues',ax=ax[0][1])
sns.barplot(data=df,x='posse_de_imovel',y='renda',palette='Blues',ax=ax[1][0])
sns.barplot(data=df,x='posse_de_veiculo',y='renda',palette='Blues',ax=ax[1][1])
plt.tight_layout()
````

````python
# Gráfico pointplot para posse de imóvel
plt.figure(figsize=(10, 6))
sns.pointplot(x='posse_de_imovel', y='renda', data=df, ci='sd',hue='sexo',palette='Blues')
plt.title('Renda média por posse de imóvel')
plt.show()

````

````python
# Gráfico pointplot para posse de veículo
plt.figure(figsize=(10, 6))
sns.pointplot(x='posse_de_veiculo', y='renda', data=df, ci='sd',hue='sexo',palette='Blues')
plt.title('Renda média por posse de veículo')
plt.show()
````
